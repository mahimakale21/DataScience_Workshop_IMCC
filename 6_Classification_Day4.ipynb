## Decision Tree

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### import dataset

```python
df = pd.read_csv('Social_Network_Ads.csv')
df.shape
```

```python
list(df.columns)
```

```python
df
```

### separate  the i/p and o/p variables

```python
x = df[['Age','EstimatedSalary']]
y = df['Purchased']
```

```python
x
```

### explore data

```python
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid()
plt.scatter(x['Age'], x['EstimatedSalary'], c = y)
```

### import the class

```python
from sklearn.tree import DecisionTreeClassifier
```

### create object

```python
classifier = DecisionTreeClassifier(random_state = 0)
```

### Train model

```python
classifier.fit(x,y)
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize = (12,15))
plot_tree(classifier, fontsize = 5, feature_names = ['age', 'sal'], class_names = ['y','n'], filled = True, rounded = True);
```

### Prediction on unseen data

```python
new = pd.DataFrame({
    'Age': [25,37,56,66],
    'EstimatedSalary': [34000,41000, 26000,134000]
})
new
```

```python
classifier.predict(new)
```
