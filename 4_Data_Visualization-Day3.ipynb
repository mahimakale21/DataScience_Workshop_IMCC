```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Bivariate-data-Visualization

```python
x = np.array([3,13,19,31,38,45,58,65,71,88])
y = np.array([11,16,24,37,41,58,56,78,81,95])
z = np.array([81,80,79,63,54,32,29,26,18,14])
```

## connectivity

```python
plt.title('Plot of X vs Y')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.grid()
plt.xlim(0,100)
plt.ylim(0,100)
plt.xticks(range(0,101,10))
plt.yticks(range(0,101,10))
plt.plot(x,y, color = 'purple', marker = 'o', ls = ':', mfc='blue')
```

```python
plt.title('Plot of X vs Z')
plt.xlabel('X values')
plt.ylabel('z values')
plt.grid()
plt.xlim(0,100)
plt.ylim(0,100)
plt.xticks(range(0,101,10))
plt.yticks(range(0,101,10))
plt.plot(x,z)
```

```python
plt.title('2 plots')
plt.xlabel('x-values')
plt.ylabel('y-values and z=values')
plt.grid()
plt.plot(x,y,color = 'red',label = '2023')
plt.plot(x,z,color = 'green',label = '2022')
plt.legend()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y)
plt.xlabel('X values')
plt.ylabel('y values')
plt.subplot(2,1,2)
plt.plot(x,z)
plt.xlabel('X values')
plt.ylabel('z values')
```

```python
plt.subplot(1,2,1) ##(row,col,plot_no)
plt.plot(x,y)
plt.xlabel('X values')
plt.ylabel('y values')
plt.subplot(122) ##can be written without commas
plt.plot(x,z)
plt.xlabel('X values')
plt.ylabel('z values')
```

## change aspect ratio using figure-size 

```python
plt.figure(figsize = (16,9))
plt.title('Plot of X vs Z')
plt.xlabel('X values')
plt.ylabel('z values')
plt.grid()
plt.xlim(0,100)
plt.ylim(0,100)
plt.xticks(range(0,101,10))
plt.yticks(range(0,101,10))
plt.plot(x,z, lw = 4, marker = 'o', ms = 15, mfc = 'pink')
plt.savefig('myplot.png')
```

## Scatter Plot

```python
plt.title('Plot of X vs Y')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.grid()
plt.scatter(x,y)
```

```python
java = np.array([43,78,65,69,63,82,47,77,85,50,66,81])
python = np.array([46,81,62,66,69,77,41,75,88,43,65,83])
cols = [1,2,3,3,3,2,1,2,2,1,3,2]
```

## Bar Plot

```python
enrollment = [104,112,256,178,126,180]
courses = ['C','Java','Python','R','PHP','JavaScript']
```

```python
plt.xlabel('Courses')
plt.ylabel('Enrolled students')
plt.grid()
plt.bar(courses,enrollment,color = 'turquoise')
```

```python
plt.ylabel('Courses')
plt.xlabel('Enrolled students')
plt.grid()
plt.barh(courses,enrollment,color = 'teal')
```

## Pie-Chart

```python
plt.figure(figsize= (8,8))
plt.pie(enrollment, labels = courses, autopct= '%2.2f', shadow = True);
```

# Univariate-data-Visualization

## Histogram

```python
marks = np.array([67,56,45,61,93,72,65,58,65,49,68,62])
```

```python
plt.title('Frequency plot of Marks')
plt.xlabel('Range of Marks')
plt.ylabel('Count')
plt.grid()
plt.xticks(range(0,101,10))
plt.hist(marks, bins = range(0,101,10),color = 'orange');
```

## Boxplot

```python
plt.boxplot(marks);
```
